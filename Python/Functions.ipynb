```python
# Scope of Variables
image_scope = Image.open(r'Scope of Variables.png')
display(image_scope)
```

```python
# Example
def scope():
    fun_var = 20
    return "Value of variable inside the function:",fun_var
```

```python
fun_var = 30
scope()
print("Value of variable outside the function:",fun_var)
```

Value of variable outside function with same name has different value of 30 becuase the function variable gets destroyed with the function whereas the global variable is not destroyed. Thus the local variable to the function is only accessed by the function and after the funtion returns a value, then we exit the function which destroys the local variable.
